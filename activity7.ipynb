<h1> Regional Analysis </h1>
<h2>In the year 1970, Most of North America has incident count with an attack type of Bombing/Explosion. We can see it clearly in the HeatMap below</h2>

```python
# Step 1
regional_filter = df[(df['iyear'] >= 1970) &
                    (df['attacktype1_txt'] == 'Bombing/Explosion' ) &
                    (df['region_txt'] == 'North America')]
aggregated_df = regional_filter.groupby(['latitude', 'longitude']).size().reset_index(name= 'incident_count')

# Step 2
aggregated_df['normalize_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                    (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

# Step 3
heat_data = aggregated_df[['latitude', 'longitude', 'incident_count']].values.tolist()

# Step 4
m = folium.Map(location=[30, 40], zoom_start=2)

# Step 5
HeatMap(heat_data).add_to(m)

# Step 6
m.save('regional_terrorism_heatmap.html')
m
```

<h1> Crime Type Analysis </h1>
<h2> For the year 2017, crime type Insurgency/Guerilla Action can be seen in the map present in philippines and part of Africa also some part of middle east </h2>

```python
#1
crimetype_filter = df[(df['iyear'] >= 2017) &
                    (df['alternative_txt'] == 'Insurgency/Guerilla Action' ) &
                    (df['nkill'] >= 5)]

#2
aggregated_df = crimetype_filter.groupby(['latitude', 'longitude']).size().reset_index(name= 'incident_count')

#3
aggregated_df ['normalize_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

aggregated_df = aggregated_df.dropna(subset=['latitude', 'longitude'])

#4
heat_data = aggregated_df[['latitude', 'longitude','normalize_weight']].values.tolist()

#5
m = folium.Map(location=[30, 40], zoom_start=4)

HeatMap(heat_data).add_to(m)

#6
m.save('terrorism_heatmap.html')
 
m
```

<h1> Casualty Analysis </h1>
<h2> There is more casualty in the mindanao region in the Philippines in terms of the explosive weapon type in the year 1997 </h2>

```python
# Step 1
casualty_filter = df[(df['iyear'] >= 1997) &
                    (df['weaptype1_txt'] == 'Explosives' ) &
                    (df['country_txt'] == 'Philippines' ) &
                    (df['nkill'] >= 1)]

aggregated_df = casualty_filter.groupby(['latitude', 'longitude']).size().reset_index(name= 'incident_count')

# Step 2
aggregated_df['normalize_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                    (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

# Step 3 
heat_data = aggregated_df[['latitude', 'longitude', 'incident_count']].values.tolist()

# Step 4
m = folium.Map(location=[30, 40], zoom_start=2)

# Step 5 
HeatMap(heat_data).add_to(m)

# Step 6 
m.save('crime_type_heatmap.html')
m
```

<h1> Demographic Analysis </h1>

<h1> Target Analysis </h1>
<h2> In the year 1970 where the record starts, part of europe has the highest incident count and the target of it is the Government (Diplomatic). There is also records from South America ans Middle East</h2>

```python
# Step 1
casualty_filter = df[(df['iyear'] >= 1970) &
                    (df['targtype1_txt'] == 'Government (Diplomatic)' ) &
                    (df['attacktype1_txt'] == 'Assassination' ) &
                    (df['success'] == 1) &
                    (df['nkill'] >= 1)]

aggregated_df = casualty_filter.groupby(['latitude', 'longitude']).size().reset_index(name= 'incident_count')

# Step 2
aggregated_df['normalize_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                    (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

aggregated_df = aggregated_df.dropna(subset=['latitude', 'longitude'])

# Step 3
heat_data = aggregated_df[['latitude', 'longitude', 'incident_count']].values.tolist()

# Step 4
m = folium.Map(location=[30, 40], zoom_start=2)

# Step 5
HeatMap(heat_data).add_to(m)

# Step 6
m.save('target_analysis_heatmap.html')
m
```
